```python
import pandas as pd
import numpy as np
data={
     'users':['steve','nancy','dustin','mike','lucas','ele','will','max'],
     'id':[12,67,11,72,21,89,72,2],
     'season':['s1','s1','s1','s1','s1','s1','s1','s2']
}
df=pd.DataFrame(data)
print('casting list of stranger things\n',df)
```

filtering

```python
df=df[df['season']=='s2']
print(df)
```

```python
df=df[df['users'].isin(['mike','will'])]
print(df)
```

grouping

```python
df.groupby('season')['users'].count()
```

```python
df.groupby('season')['id'].mean()
```

```python
df.groupby('season')['id'].max()
```

```python
df.groupby('season')['id'].sum()
```

```python
df.groupby('id')['users'].apply(list)
```

**data type casting**

```python
df['id'] = pd.to_numeric(df['id'], errors='coerce')
df['season'] = df['season'].astype('category')
df['users'] = df['users'].astype('category')
```

```python
print(df)
```

**Future engineering**
*`Feature engineering means creating, transforming, or selecting features (columns) to improve model performance. Since your DataFrame includes users, id, and season, lets explore meaningful ways to engineer features from this data.`*

```python
df['season_encoded'] = df['season'].map({'s1': 0, 's2': 1})
```

```python
season_dummies = pd.get_dummies(df['season'], prefix='season')
df = pd.concat([df, season_dummies], axis=1)
```

```python
print(df)
```

```python
user_freq = df['users'].value_counts().to_dict()
df['user_freq'] = df['users'].map(user_freq)
```

```python
print(season_dummies,'\n')
print(user_freq)
```

```python
df['is_id_duplicated'] = df.duplicated('id', keep=False).astype(int)
```

```python
df.groupby('season')['id'].agg(['mean', 'max', 'min'])
```

```python
main_users = ['mike', 'will', 'nancy']
df['is_main'] = df['users'].isin(main_users).astype(int)
```
